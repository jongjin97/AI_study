# SemanticChunker 생성

```python
from dotenv import load_dotenv

# API 키 정보 로드
load_dotenv()
```

```python
# data/appendix-keywords.txt 파일을 열어서 f라는 파일 객체를 생성합니다.
with open("./data/appendix-keywords.txt", encoding='utf-8') as f:
    file = f.read()  # 파일의 내용을 읽어서 file 변수에 저장합니다.

# 파일으로부터 읽은 내용을 일부 출력합니다.
print(file[:350])
```

```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai import OpenAIEmbeddings

# OpenAI 임베딩을 사용하여 의미론적 청크 분할기를 초기화
text_splitter = SemanticChunker(embeddings=OpenAIEmbeddings())
```

# 텍스트 분할

```python
chunks = text_splitter.split_text(file)
```

```python
# 분할된 청크 중 첫 번째 청크를 출력합니다.
print(chunks[0])
```

```python
# text_splitter를 사용하여 분할
docs = text_splitter.create_documents([file])
print(docs[0].page_content)
```

# Breakpoints

```python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_type="percentile",
    breakpoint_threshold_amount=70,
)
```

```python
docs = text_splitter.create_documents([file])
for i, doc in enumerate(docs[:5]):
    print(f"[Chunk {i}]", end="\n\n")
    print(doc.page_content)
    print("===" * 20)
```

```python
print(len(docs))
```

```python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_amount=1.25,
    breakpoint_threshold_type="standard_deviation"
)
```

```python
# text_splitter를 사용하여 분할
docs = text_splitter.create_documents([file])
```

```python
for i, doc in enumerate(docs[:5]):
    print(f"[Chunk {i}]", end="\n\n")
    print(doc.page_content)
    print("===" * 20)
```

```python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(),
    breakpoint_threshold_amount=0.5,
    breakpoint_threshold_type="interquartile"
)
```

```python
docs = text_splitter.create_documents([file])

for i, doc in enumerate(docs[:5]):
    print(f"[Chunk {i}]", end="\n\n")
    print(doc.page_content)
    print("===" * 20)
```
